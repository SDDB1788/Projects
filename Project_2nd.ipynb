# Q8.To find 100 sneakers in Flipkart

```python
driver=webdriver.Chrome()
```

```python
driver.get("https://www.flipkart.com/")
```

```python
search_tag=driver.find_element_by_xpath("//input[@class='_3704LK']")
search_tag.send_keys('sneakers')
search_btn=driver.find_element_by_xpath("//button[@class='L0Z3Pu']")
search_btn.click()
```

```python
brand=[]
product_des=[]
price=[]
Discount_percent=[]
for d in range(0,3):
    for i in driver.find_elements_by_xpath("//div[@class='_2WkVRV']"):
        brand.append(i.text)
    for x in driver.find_elements_by_xpath("//a[@class='IRpwTa']"):
        product_des.append(x.text)
    for e in driver.find_elements_by_xpath("//div[@class='_30jeq3']"):
        price.append(e.text)
    for w in driver.find_elements_by_xpath("//div[@class='_3Ay6Sb']/span[1]"):
        Discount_percent.append(w.text)
    time.sleep(2)
    url=driver.find_element_by_xpath("//a[@class='_1LKTO3']").get_attribute("href")
    driver.get(url)

    
    
    
```

```python
len(brand),'',len(product_des),'',len(price),'',len(Discount_percent)
```

```python
#why there is a difference in the length of each field?
```

```python
df=pd.DataFrame({})
df['Brand']=brand[:101]
df['Product_Des']=product_des[:101]
df['Price']=price[:101]
df['Discount_Percent']=Discount_percent[:101]

df
```

# 10. Amazon- Laptop

```python
driver.get("https://www.amazon.in/")
```

```python
search_tag=driver.find_element_by_xpath("//input[@id='twotabsearchtextbox']")
search_tag.send_keys("Laptop")
search_btn=driver.find_element_by_xpath("//input[@id='nav-search-submit-button']")
search_btn.click()
```

#to set criteria for i7 and i9

```python
filter_btn=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_btn:
    if i.text=='Intel Core i7':
        i.click()
        break
        
    
```

```python
filter_btn=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_btn:
    if i.text=='Intel Core i9':
        i.click()
        break
```

to scrap titles,price and rating

```python
Title=[]
Price=[]

Titles=driver.find_elements_by_xpath("//span[@class='a-size-medium a-color-base a-text-normal']")
for i in Titles[:10]:
    Title.append(i.text)
Prices=driver.find_elements_by_xpath("//span[@class='a-price-whole']")
for o in Prices[:10]:
    Price.append(o.text)
    
```

```python
Ratings=[]

url=driver.find_elements_by_xpath("//a[@class='a-link-normal a-text-normal']")
Ur=[]
for i in url[:10]:
    Ur.append(i.get_attribute('href'))
for url in Ur:
    driver.get(url)
    try:
        rate=driver.find_element_by_xpath("//span[@id='acrCustomerReviewText']")
        rate.click()
        rating=driver.find_element_by_xpath("//span[@class='a-size-medium a-color-base']")
        Ratings.append(rating.text)
    except NoSuchElementException as e:
        Ratings.append("No Rating")
```

```python
len(Title),'',len(Price),'',len(Ratings)
```

```python
Ratings
```

```python
df=pd.DataFrame({'Title':Title,
                'Price':Price,
                'Ratings':Ratings})

df
```
