```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
lg=LogisticRegression()
```

```python
#LogisticRegression used as y is only two(edible or non-edible(yes/no)) and repsective matrix also called for the test
```

```python
from sklearn.model_selection import train_test_split
for i in range(0,1000):
    x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=i)
    lg.fit(x_train,y_train)
    pred_train=lg.predict(x_train)
    pred_test=lg.predict(x_test)
    if round(accuracy_score(y_train,pred_train)*100,2)==round(accuracy_score(y_test,pred_test)*100,2):
        print("At random state",i,"The model performs very well")
        print("At random state:-",i)
        print("Training accuracy_score:-",accuracy_score(y_train,pred_train)*100)
        print("Testing accuracy_score:-",accuracy_score(y_test,pred_test)*100)
```
